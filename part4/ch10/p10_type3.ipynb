### 문제1-1

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bae-v3/main/part4/ch10/attrition.csv")

# 1) 로지스틱 회귀분석 수행
from statsmodels.formula.api import logit
model = logit("attrition ~ age + income + overtime", data=df).fit()

# 2) 회귀 분석 결과 출력
print(model.summary())

# 3) income 회귀계수 출력
print(round(model.params["income"], 3))

# 정답: -0.002
```

### 문제1-2

```python
# 1) 오즈비 계산
import numpy as np
result = np.exp(model.params["age"])

# 2) 반올림 후 결과 출력
print(round(result, 3))

# 정답: 0.996
```

### 문제1-3

```python
# 1) 새로운 직원 데이터 만들기
new = pd.DataFrame({"age":[40], "income":[4500], "overtime":[1]})

# 2) 예측
pred = model.predict(new)

# 3) 반올림 후 결과 출력
print(round(pred[0], 3))
```

### 문제2-1

```python
import pandas as pd

# 1) 데이터 불러오기
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bae-v3/main/part4/ch10/heating.csv")

# 2) 다중선형회귀분석
from statsmodels.formula.api import ols
model = ols("heating_load ~ wall + roof + glazing + height", data=df).fit()

# 3) 회귀 분석 결과 출력
print(model.summary())

# 4) 절편 제외 회귀계수들의 합 구하기
result = model.params[1:].sum()
print(round(result, 3))

# 정답: 0.254
```

### 문제2-2

```python
# 1) 다시 회귀 분석
model = ols("heating_load ~ roof + glazing + height", data=df).fit()

# 2) 결정계수(R-squared) 확인
print(model.summary())

# 정답: 0.754
```

### 문제2-3

```python
# 1) 새로운 데이터 (예: 문제에서 제시된 값)
new = pd.DataFrame({"wall":[20],"roof":[150],"glazing":[20],"height":[5]})

# 2) 예측값 계산
pred = model.predict(new)

# 3) 반올림 후 출력
print(round(pred[0], 3))

# 정답: 79.612
```
